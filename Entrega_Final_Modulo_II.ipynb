## Función generadora de momentos (FGM)

Debe incluir definición, propiedades, obtención de momentos y al menos un ejemplo.

$$
M_X(t)=\mathbb{E}(e^{tX})
$$

Los momentos se obtienen derivando:

$$
\mathbb{E}(X)=M_X'(0), \qquad \mathbb{E}(X^2)=M_X''(0).
$$

## Función característica (FC)

Debe incluir definición, propiedades, relación con los momentos y al menos un ejemplo.

$$
\varphi_X(t)=\mathbb{E}(e^{itX})
$$

La función característica siempre existe y caracteriza completamente la distribución.
